```python
# ==============================
# PHISHING EMAIL ANALYSIS PROJECT
# ==============================

# Import required libraries
import re
import pandas as pd

print("🚀 Starting Phishing Email Analysis...\n")

# ==============================
# STEP 1: SAMPLE EMAIL INPUT
# ==============================

email_text = """
Subject: Urgent! Your account has been suspended

Dear User,

We noticed suspicious activity in your account. Please verify immediately.

Click here: http://fakebank-login.com

If you do not verify, your account will be permanently suspended.

Thank you,
Security Team
"""

print("📧 Email Loaded Successfully!\n")


# ==============================
# STEP 2: EXTRACT URLS
# ==============================

urls = re.findall(r'https?://\S+', email_text)

print("🔗 Extracted URLs:")
print(urls if urls else "No URLs found")
print()


# ==============================
# STEP 3: DETECT SUSPICIOUS KEYWORDS
# ==============================

keywords = ["urgent", "verify", "suspended", "click", "immediately", "account", "security"]

found_keywords = [word for word in keywords if word in email_text.lower()]

print("⚠️ Suspicious Keywords Found:")
print(found_keywords if found_keywords else "No suspicious keywords")
print()


# ==============================
# STEP 4: RISK SCORING
# ==============================

score = 0

if urls:
    score += 2

if len(found_keywords) > 2:
    score += 2

print("📊 Risk Score:", score)

if score >= 3:
    result = "PHISHING"
    print("🚨 This email is likely PHISHING\n")
else:
    result = "SAFE"
    print("✅ This email seems SAFE\n")


# ==============================
# STEP 5: CREATE DATAFRAME
# ==============================

data = {
    "URLs Found": [len(urls)],
    "Keywords Found": [len(found_keywords)],
    "Risk Score": [score],
    "Final Result": [result]
}

df = pd.DataFrame(data)

print("📋 Analysis Summary Table:")
print(df)
print()


# ==============================
# STEP 6: SAVE RESULTS TO FILE
# ==============================

# Save CSV
df.to_csv("analysis_output.csv", index=False)

# Save sample email
with open("sample_email.txt", "w") as f:
    f.write(email_text)

print("💾 Files Saved Successfully:")
print("- analysis_output.csv")
print("- sample_email.txt\n")


# ==============================
# STEP 7: GENERATE REPORT TEXT FILE
# ==============================

report_text = f"""
PHISHING EMAIL ANALYSIS REPORT

Objective:
To analyze suspicious email content and identify phishing indicators.

Tools Used:
- Python
- Jupyter Notebook
- Regex
- Pandas

Analysis Approach:
- Extract URLs
- Detect suspicious keywords
- Assign risk score

Findings:
- URLs Found: {len(urls)}
- Keywords Found: {len(found_keywords)}
- Risk Score: {score}
- Final Result: {result}

Conclusion:
This email is classified as {result} based on detected phishing indicators.
"""

with open("phishing_report.txt", "w") as f:
    f.write(report_text)

print("📄 Report Generated: phishing_report.txt\n")


# ==============================
# FINAL MESSAGE
# ==============================

print("🎉 Project Completed Successfully!")
print("You can now upload files to GitHub 🚀")
```
